```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/nehag/OneDrive/Desktop/project/Data/raw/hotel_bookings.csv")
df.head()

df.shape
df.columns
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# fill missing values
df.fillna({
    'children': 0,
    'country': 'Unknown',
    'agent': 0,
    'company': 0
}, inplace=True)
```

```python
# remove duplicates
df = df.drop_duplicates()
```

```python
# Remove extreme ADR values
df = df[df['adr'] < 500]
```

```python
# remove rows with no guests
df = df[(df['adults'] + df['children'] + df['babies']) > 0]
```

```python
df['is_family'] = df['children'] + df['babies'] > 0
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
def lead_category(x):
    if x < 7:
        return 'Last-Minute'
    elif x < 30:
        return 'Mid-Term'
    else:
        return 'Early'

df['lead_time_category'] = df['lead_time'].apply(lead_category)
```

```python
df.info()
df.isnull().sum()
print("Final dataset shape:", df.shape)
```

```python
df.to_csv("../Data/processed/cleaned_hotel_bookings.csv", index=False)
```
